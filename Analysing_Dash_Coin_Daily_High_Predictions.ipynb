```python
exp = []
pred = []
f = open("predictions.txt", "r")
for i in f:
    v = list(map(float, i.split()))
    exp.append(v[0])
    pred.append(v[1])
```

```python
print(len(pred))
```

```python
import scipy.stats as sp
correl = sp.pearsonr(exp, pred)
print(correl)
```

```python
import matplotlib.pyplot as plt
plt.plot(exp)
plt.plot(pred)
plt.ylabel("price change")
plt.show()
```
